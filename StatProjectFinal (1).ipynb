```python
import pandas as pd

# Replace 'your_file.csv' with the actual name of your CSV file
file_path = 'datacategories.csv'
df = pd.read_csv(file_path)
print(df.head()) # Display the first few rows of the DataFrame
```

```python
df.columns = df.columns.str.strip()
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the histogram of the numeric GPA column
sns.histplot(df['GPA'], kde=True, bins=10, color='blue')
plt.title('Original GPA Distribution')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['InternetUsageTime'], kde=True, bins=10, color='blue')
plt.title('Internet Usage Distribution')
plt.xlabel('InternetUsageTime')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['GenerativeAIUsageFrequency'], kde=True, bins=10, color='blue')
plt.title('GenerativeAIUsage')
plt.xlabel('GenerativeAIUsageFrequency')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['SNSUsage'], kde=True, bins=10, color='blue')
plt.title('Social Networking Sites Usage Distribution')
plt.xlabel('SNSUsageTime')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import chi2_contingency



# Create a contingency table
contingency_table = pd.crosstab(df['InternetUsageTime'], df['GPA'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

```python
# Plot observed frequencies
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=False, fmt="d", cmap="Blues")
plt.title("Observed Frequencies")
plt.xlabel("GPA")
plt.ylabel("InternetUsagetime")
plt.show()

# Plot expected frequencies
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)

plt.figure(figsize=(8, 6))
sns.heatmap(expected_df, annot=False, fmt=".2f", cmap="Greens")
plt.title("Expected Frequencies")
plt.xlabel("GPA")
plt.ylabel("InternetUsagetime")
plt.show()
```

```python
# Plot stacked bar chart
contingency_table.plot(kind='bar', stacked=True, colormap='viridis', figsize=(8, 6))
plt.title('Stacked Bar Chart')
plt.ylabel('Counts')
plt.xlabel('InternetUsagetime')
plt.show()

```
